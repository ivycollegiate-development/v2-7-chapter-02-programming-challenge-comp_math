```python
# imports matpolotlib.pylot library for data visualization and numpy library for mathematical operations
import matplotlib.pyplot as plt
import numpy as np

# Function to generate a range of numbers
# parameters: start: starting number, final: final value, increment: value to increase each time
def frange(start, final, increment):
    numbers = []
    while start < final:
        numbers.append(start)
        start += increment
    return numbers

# Function to compare multiple quatratic functions
# puts multiple quatratic functions on one graph for comparison
def compare_quadratic():
    x_values = frange(-1, 5, 0.1) # Generates x-values
    y_list = [
        lambda x: x**2 + 1,
        lambda x: x**2 - 1,
        lambda x: x**2 / 2,
        lambda x: x**2 * 2
    ]

# Generate the X Values using the frange fucntion
#    x_values = [-1, 1, 2, 3, 4, 5]
    labels = [
        'y = x² + 1',
        'y = x² - 1',
        'y = x² / 2',
        'y = x² * 2',
    ]

    # Plot each quadratic function
    for y_func, label in zip(y_list, labels):
        y_values = [y_func(x) for x in x_values]  # Calculate y-values for each function
        plt.plot(x_values, y_values, label=label)

# Adds legend and grid    
    plt.legend(f'{a}'for a in y_list)
    plt.grid(True, linestyle='--', alpha=0.7)
    # Adds x and y labels
    plt.xlabel('X Values')
    plt.ylabel('Y Values (X²)')
    # Adds title before plotting
    plt.title('Explore Quadratic Function')
    plt.show()


'''
    for a in y_list:
        draw_quadratic(x_values,a)

#Function to plot the calculated quatratic functions
def draw_quadratic(x,y):
    x=[]
    y=[]
    intervals = frange(-20, 20, 0.01)
    for b in intervals:
        x.append(b)
    draw_graph(x,y)

def draw_graph (j, c):
    plt.plot(j, c,'b' ,label='Quadratic(y=x^2)')
#   plt.xlabel('X Values')
#   plt.ylabel('Y Values (X²)')
    plt.title('Explore Quadratic Function')
'''
# Call the function to display the graph
compare_quadratic()
```
